```python
import pandas as pd
import re

# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏ –Ω–∞–ª–∏—á–∏—è —ç–º–æ–¥–∑–∏
def contains_emoji(text):
    emoji_pattern = re.compile("[\U0001F600-\U0001F64F]"  # –°–º–∞–π–ª—ã
                               "|[\U0001F300-\U0001F5FF]"  # –°–∏–º–≤–æ–ª—ã & –ø–∏–∫—Ç–æ–≥—Ä–∞–º–º—ã
                               "|[\U0001F680-\U0001F6FF]"  # –¢—Ä–∞–Ω—Å–ø–æ—Ä—Ç & —Å–∏–º–≤–æ–ª—ã –∫–∞—Ä—Ç
                               "|[\U0001F700-\U0001F77F]"  # –ê–ª—Ö–∏–º–∏—á–µ—Å–∫–∏–µ —Å–∏–º–≤–æ–ª—ã
                               "|[\U0001F780-\U0001F7FF]"  # –ì–µ–æ–º–µ—Ç—Ä–∏—á–µ—Å–∫–∏–µ —Ñ–∏–≥—É—Ä—ã
                               "|[\U0001F800-\U0001F8FF]"  # –î–æ–ø. —Å–∏–º–≤–æ–ª—ã
                               "|[\U0001F900-\U0001F9FF]"  # –†—É–∫–æ–ø–æ–∂–∞—Ç–∏—è –∏ —ç–º–æ—Ü–∏–∏
                               "|[\U0001FA00-\U0001FA6F]"  # –°–∏–º–≤–æ–ª—ã –æ–±—ä–µ–∫—Ç–æ–≤
                               "|[\U0001FA70-\U0001FAFF]"  # –î–æ–ø. —Å–∏–º–≤–æ–ª—ã –æ–±—ä–µ–∫—Ç–æ–≤
                               "|[\U00002702-\U000027B0]"  # –†–∞–∑–Ω—ã–µ —Å–∏–º–≤–æ–ª—ã
                               "|[\U000024C2-\U0001F251]"  # –†–∞–∑–Ω–æ–µ
                               , flags=re.UNICODE)
    return 1 if emoji_pattern.search(str(text)) else 0

df['emoji'] = df['about'].apply(contains_emoji)

df
```

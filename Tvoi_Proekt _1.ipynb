### MBTI 

```python
# –í—ã–±–æ—Ä –Ω—É–∂–Ω—ã—Ö —Å—Ç–æ–ª–±—Ü–æ–≤
selected_columns = ["ENFJ", "ENFP", "ENTJ", "ENTP", "ESFJ", "ESFP",
                    "ESTJ", "ESTP", "INFJ", "INFP", "INTJ", "INTP",
                    "ISFJ", "ISFP", "ISTJ", "ISTP"]
MBTI = df[selected_columns]

print("–ù–æ–≤—ã–π –¥–∞—Ç–∞—Ñ—Ä–µ–π–º MBTI: ")
MBTI
```

```python
# –ü–æ–¥—Å—á–µ—Ç –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –∑–Ω–∞—á–µ–Ω–∏–π 1.0 –≤ –∫–∞–∂–¥–æ–º —Å—Ç–æ–ª–±—Ü–µ
count_ones = (MBTI == 1.0).sum()

print("–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–Ω–∞—á–µ–Ω–∏–π 1.0 –≤ –∫–∞–∂–¥–æ–º —Å—Ç–æ–ª–±—Ü–µ:")
count_ones
```

```python

# –í—ã–≤–æ–¥–∏–º –¥–∞—Ç–∞—Ñ—Ä–µ–π–º
df


# –ê–Ω–∞–ª–∏–∑ —É—Å–ø–µ—à–Ω–æ—Å—Ç–∏ –≤–∑–∞–∏–º–æ–¥–µ–π—Å—Ç–≤–∏–π
print("\n–°—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ª–∞–π–∫–æ–≤, –ø–æ–ª—É—á–µ–Ω–Ω—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è–º–∏:")
print(df["likes_received_count"].median())

print("\n–°—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ª–∞–π–∫–æ–≤, –ø–æ—Å—Ç–∞–≤–ª–µ–Ω–Ω—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è–º–∏:")
print(df["likes_given_count"].median())

print("\n–°—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –º—ç—Ç—á–µ–π:")
print(df["matches_count"].median())

print("\n–ö–æ—Ä—Ä–µ–ª—è—Ü–∏—è –º–µ–∂–¥—É –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º –ª–∞–π–∫–æ–≤ –∏ –º—ç—Ç—á–∞–º–∏:")
print(df[["likes_received_count", "matches_count"]].corr())

print("\n–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π —Å –ø–æ–¥–ø–∏—Å–∫–æ–π Premium:")
print(df["is_premium"].sum())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# –°–æ–∑–¥–∞–µ–º –¥–∞—Ç–∞—Ñ—Ä–µ–π–º —Å —Å–æ—Ü–∏–∞–ª—å–Ω–æ-–¥–µ–º–æ–≥—Ä–∞—Ñ–∏—á–µ—Å–∫–∏–º–∏ —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫–∞–º–∏ –∏ –∞–∫—Ç–∏–≤–Ω–æ—Å—Ç—å—é


dff = df

# –í—ã–≤–æ–¥–∏–º –º–µ–¥–∏–∞–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è
print("\n–ú–µ–¥–∏–∞–Ω–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ª–∞–π–∫–æ–≤, –ø–æ–ª—É—á–µ–Ω–Ω—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è–º–∏:")
print(dff["likes_received_count"].median())

print("\n–ú–µ–¥–∏–∞–Ω–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ª–∞–π–∫–æ–≤, –ø–æ—Å—Ç–∞–≤–ª–µ–Ω–Ω—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è–º–∏:")
print(dff["likes_given_count"].median())

print("\n–ú–µ–¥–∏–∞–Ω–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –º—ç—Ç—á–µ–π:")
print(dff["matches_count"].median())

# –ì—Ä–∞—Ñ–∏–∫ 1: –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ª–∞–π–∫–æ–≤, –ø–æ–ª—É—á–µ–Ω–Ω—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è–º–∏
plt.figure(figsize=(8, 5))
sns.histplot(dff["likes_received_count"], bins=10, kde=True, color="blue")
plt.axvline(dff["likes_received_count"].median(), color="red", linestyle="--", label="–ú–µ–¥–∏–∞–Ω–∞")
plt.title("–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ–ª—É—á–µ–Ω–Ω—ã—Ö –ª–∞–π–∫–æ–≤")
plt.xlabel("–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ª–∞–π–∫–æ–≤")
plt.ylabel("–ß–∞—Å—Ç–æ—Ç–∞")
plt.legend()
plt.show()

# –ì—Ä–∞—Ñ–∏–∫ 2: –ú–µ–¥–∏–∞–Ω–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ª–∞–π–∫–æ–≤ –ø–æ –ø–æ–ª—É
plt.figure(figsize=(8, 5))
sns.barplot(x="gender", y="likes_received_count", data=dff, estimator=pd.Series.median, palette="Set2")
plt.title("–ú–µ–¥–∏–∞–Ω–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ª–∞–π–∫–æ–≤ –ø–æ –ø–æ–ª—É")
plt.xlabel("–ü–æ–ª")
plt.ylabel("–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ª–∞–π–∫–æ–≤")
plt.show()

# –ì—Ä–∞—Ñ–∏–∫ 3: –ö–æ—Ä—Ä–µ–ª—è—Ü–∏—è –ª–∞–π–∫–æ–≤ –∏ –º—ç—Ç—á–µ–π
plt.figure(figsize=(8, 5))
sns.scatterplot(x="likes_received_count", y="matches_count", hue="gender", data=dff)
plt.title("–ö–æ—Ä—Ä–µ–ª—è—Ü–∏—è –ø–æ–ª—É—á–µ–Ω–Ω—ã—Ö –ª–∞–π–∫–æ–≤ –∏ –º—ç—Ç—á–µ–π")
plt.xlabel("–ü–æ–ª—É—á–µ–Ω–Ω—ã–µ –ª–∞–π–∫–∏")
plt.ylabel("–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –º—ç—Ç—á–µ–π")
plt.show()

# –ì—Ä–∞—Ñ–∏–∫ 4: –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–µ–º–∏—É–º-–ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π
plt.figure(figsize=(6, 4))
premium_counts = dff["is_premium"].value_counts()
sns.barplot(x=premium_counts.index, y=premium_counts.values, palette="pastel")
plt.title("–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π —Å –ø–æ–¥–ø–∏—Å–∫–æ–π Premium")
plt.xlabel("–ü–æ–¥–ø–∏—Å–∫–∞ Premium")
plt.ylabel("–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π")
plt.xticks([0, 1], ["–ù–µ—Ç", "–î–∞"])
plt.show()
```
